```python
import oracle 
import sys
import numpy as np
import matplotlib.pyplot as plt
import os

X_train, y_train, X_test, y_test = oracle.q3_linear_1(23746)
X_train = np.array(X_train)
y_train = np.array(y_train)
X_test = np.array(X_test)
y_test = np.array(y_test)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# we will run a linear regression model on the data

def linear_regression(X, y):
    w = np.linalg.inv(X.T @ X) @ X.T @ y
    return w

def predict(X, w):
    return X @ w

def mse(y, y_pred):
    return np.mean((y - y_pred)**2)

w = linear_regression(X_train, y_train)
print(w)
y_pred = predict(X_test, w)
print("Test data :",mse(y_test, y_pred))
y_pred = predict(X_train, w)
print("Train data :",mse(y_train, y_pred))


```

```python
def Ridge_regression(X, y, alpha):
    # print(np.eye(X.shape[0]))
    w = np.linalg.inv(X.T @ X + alpha*np.eye(X.shape[1])) @ X.T @ y
    return w

def predict(X, w):
    return X @ w

def mse(y, y_pred):
    return np.mean((y - y_pred)**2)

w = Ridge_regression(X_train, y_train, 1)
print(w)

y_pred = predict(X_test, w)
print("Test data :",mse(y_test, y_pred))
y_pred = predict(X_train, w)
print("Train data :",mse(y_train, y_pred))

```

```python
import csv
import numpy as np
from sklearn.preprocessing import StandardScaler
from cvxopt import matrix, solvers
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

def load_data(stock_symbol):
    file_path = f'stocknet-dataset/price/raw/{stock_symbol}.csv'
    try:
        closing_prices = []
        with open(file_path, 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                closing_prices.append(float(row['Close']))
        return np.array(closing_prices)
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}. Please ensure the dataset is downloaded and the path is correct.")
        return None

def prepare_data(closing_prices, t):
    closing_prices_np = closing_prices.reshape(-1, 1)
    scaler = StandardScaler()
    normalized_prices = scaler.fit_transform(closing_prices_np)
    d = normalized_prices.flatten()

    N = len(d)
    X = np.array([d[i:i + t] for i in range(N - t)])
    y = d[t:]

    split_index = len(X) // 2
    X_train, X_test = X[:split_index], X[split_index:]
    y_train, y_test = y[:split_index], y[split_index:]

    return X_train, y_train, X_test, y_test, scaler

def linear_svr(X_train, y_train, C=1.0, epsilon=0.1):
    N, t = X_train.shape
    K = X_train @ X_train.T
    P = matrix(K + np.eye(N) * 1e-6)
    q = matrix(-y_train)
    G = matrix(np.vstack([-np.eye(N), np.eye(N)]))
    h = matrix(np.hstack([C * np.ones(N), np.zeros(N)]))
    sol = solvers.qp(P, q, G, h)
    alpha = np.ravel(sol['x'])
    w = np.sum(alpha[:, None] * X_train, axis=0)
    b = np.mean(y_train - X_train @ w)
    return w, b

def rbf_kernel(X1, X2, gamma):
    return np.exp(-gamma * cdist(X1, X2, 'sqeuclidean'))

def kernel_svr(X_train, y_train, gamma, C=1.0, epsilon=0.1):
    N = X_train.shape[0]
    K = rbf_kernel(X_train, X_train, gamma)
    P = matrix(K + np.eye(N) * 1e-6)
    q = matrix(-y_train)
    G = matrix(np.vstack([-np.eye(N), np.eye(N)]))
    h = matrix(np.hstack([C * np.ones(N), np.zeros(N)]))
    sol = solvers.qp(P, q, G, h)
    alpha = np.ravel(sol['x'])
    return alpha

def plot_predictions_linear(X_test, y_test, y_pred, t, gamma = None):
    avg_prices = np.mean(X_test, axis=1)
    plt.figure(figsize=(10, 5))
    plt.plot(y_test, label='Actual Closing Price')
    plt.plot(y_pred, label='Predicted Closing Price')
    plt.plot(avg_prices, label=f'Average of last {t} days')
    plt.legend()
    plt.title(f'Linear SVR Predictions for t={t}')  
    plt.show()

def plot_predictions_kernel(X_test, y_test, y_pred, t, gamma = None):
    avg_prices = np.mean(X_test, axis=1)
    plt.figure(figsize=(10, 5))
    plt.plot(y_test, label='Actual Closing Price')
    plt.plot(y_pred, label='Predicted Closing Price')
    plt.plot(avg_prices, label=f'Average of last {t} days')
    plt.legend()
    plt.title(f'SVR Predictions for t={t}, gamma={gamma}')  
    plt.show()

def main(stock_symbol):
    ts = [7, 30, 90]
    gammas = [1, 0.1, 0.01, 0.001]
    closing_prices = load_data(stock_symbol)
    if closing_prices is None:
        return
    
    for t in ts:
        X_train, y_train, X_test, y_test, scaler = prepare_data(closing_prices, t)
        
        w, b = linear_svr(X_train, y_train)
        y_pred_linear = X_test @ w + b
        plot_predictions_linear(X_test, y_test, y_pred_linear, t)
        
        for gamma in gammas:
            alpha = kernel_svr(X_train, y_train, gamma)
            K_test = rbf_kernel(X_test, X_train, gamma)
            y_pred_kernel = K_test @ alpha
            plot_predictions_kernel(X_test, y_test, y_pred_kernel, t,gamma)


if __name__ == "__main__":
    main(oracle.q3_stocknet(23746))
```
